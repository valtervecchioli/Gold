```python
import pandas as pd

df = pd.read_excel('Prices.xlsx', sheet_name='Yearly_Avg', header= 5)
df
```

```python
df.drop(columns= ['Unnamed: 0', 'Unnamed: 1'], inplace=True)
```

```python
df.rename(columns= {'Unnamed: 2': 'date'}, inplace=True)
```

```python
df.drop([x for x in range(32)], inplace=True)
```

```python
year = df['date'].dt.year
df.drop(columns= 'date', inplace=True)
```

```python
df.insert(loc=0, column='year', value= year)
df.reset_index(inplace=True)
```

```python
[x for x in df.columns if not (x=='year')]
```

```python
gold_prices = pd.melt(df, id_vars=['year'], value_vars= [x for x in df.columns if not (x=='year')])
gold_prices
```

```python
gold_prices.rename(columns={'variable': 'currency'}, inplace=True)
```

```python
gold_prices
```

```python
gold_prices.to_csv('gold_prices.csv')
```
