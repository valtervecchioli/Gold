```python
import pandas as pd

df = pd.read_excel(r'C:\Users\Lara\Documents\Ironhack\Course\Mini Projects\Mini project 4\gold_supply.xlsx')
df.head()
```

```python
df.rename(columns= {'Unnamed: 0': 'year'}, inplace=True)
df
```

```python
df_supply = df.T
df_supply
```

```python
df_supply.rename(columns= {0:'Mine production',1:'Net producer hedging', 2:'Recycled gold',3:'Total supply'}, inplace=True)
df_supply.reset_index()
```

```python
df_supply_1 = df_supply.T
```

```python
df_supply_1.drop(columns = ["year"], inplace = True)
```

```python
df_supply_2 = df_supply_1.T
df_supply_2
```

```python
df_supply_2.reset_index(inplace = True)
```

```python
df_supply_2.rename(columns= {'index':'year'}, inplace=True)
df_supply_2
```

```python
df_supply_2.columns = [col.lower() for col in df_supply_2.columns]
df_supply_2.columns = [col.replace(" ","_") for col in df_supply_2.columns]
df_supply_2
```

```python
df_supply_2.isnull().sum()
```
