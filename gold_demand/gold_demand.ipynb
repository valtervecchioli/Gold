```python
df_demand_2 = df_demand.T
```

```python
df_demand_2.drop(columns = ["year"], inplace = True)
```

```python
df_demand_3 = df_demand_2.T
```

```python
df_demand_3.reset_index(inplace = True)
```

```python
df_demand_3.rename(columns= {'index':'year'}, inplace=True)
```

```python
df_demand_3.columns = [col.lower() for col in df_demand_3.columns]
df_demand_3.columns = [col.replace(" ","_") for col in df_demand_3.columns]
df_demand_3
```

```python
df_demand_3.dtypes
```

```python
df_demand_3 = df_demand_3.astype(float)
```

```python
round_demand = df_demand_3.round(decimals=2)
round_demand
```

```python
round_demand['year'] = round_demand['year'].astype(int)
round_demand
```

```python
round_demand.to_csv("gold_demand.csv")
```
